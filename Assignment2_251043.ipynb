```python
# Q2 :

def find_dominant_strategies(matrix, row_labels=None, col_labels=None):
    num_rows = len(matrix)
    num_cols = len(matrix[0]) if num_rows > 0 else 0
    if row_labels is None: row_labels = [f"Row {i}" for i in range(num_rows)]
    if col_labels is None: col_labels = [f"Col {j}" for j in range(num_cols)]

    dominant_p1 = None
    dominant_p2 = None

    for r in range(num_rows):
        is_dominant = True
        for other_r in range(num_rows):
            if r == other_r: continue
            if not all(matrix[r][c][0] > matrix[other_r][c][0] for c in range(num_cols)):
                is_dominant = False
                break
        if is_dominant:
            dominant_p1 = r
            break

    for c in range(num_cols):
        is_dominant = True
        for other_c in range(num_cols):
            if c == other_c: continue
            if not all(matrix[r][c][1] > matrix[r][other_c][1] for r in range(num_rows)):
                is_dominant = False
                break
        if is_dominant:
            dominant_p2 = c
            break

    if dominant_p1 is not None:
        print(f"Player 1 has a strictly dominant strategy: {row_labels[dominant_p1]}")
    else:
        print("Player 1 has no strictly dominant strategy.")

    if dominant_p2 is not None:
        print(f"Player 2 has a strictly dominant strategy: {col_labels[dominant_p2]}")
    else:
        print("Player 2 has no strictly dominant strategy.")
    return dominant_p1, dominant_p2

print("(i) Advertise / Not-Advertise Matrix :")
find_dominant_strategies([[(6, 4), (8, 1)], [(4, 6), (7, 3)]], ["Advertise", "Not Advertise"], ["Advertise", "Not Advertise"])

print("\n(ii) Prisoner's Dilemma Matrix :")
find_dominant_strategies([[(-1, -1), (-3, 0)], [(0, -3), (-2, -2)]], ["Cooperate", "Defect"], ["Cooperate", "Defect"])

print("\n(iii) Stag Hunt Matrix :")
find_dominant_strategies([[(4, 4), (0, 2)], [(2, 0), (2, 2)]], ["Stag", "Hare"], ["Stag", "Hare"])
```

```python
# Q3 :

def iesds_solver(matrix, row_labels=None, col_labels=None):
    num_rows = len(matrix)
    num_cols = len(matrix[0]) if num_rows > 0 else 0
    surviving_rows = list(range(num_rows))
    surviving_cols = list(range(num_cols))

    if row_labels is None: row_labels = [f"Row {i}" for i in range(num_rows)]
    if col_labels is None: col_labels = [f"Col {j}" for j in range(num_cols)]

    elimination_made = True
    while elimination_made:
        elimination_made = False
        dominated_row = None
        for r1 in surviving_rows:
            for r2 in surviving_rows:
                if r1 != r2 and all(matrix[r1][c][0] < matrix[r2][c][0] for c in surviving_cols):
                    dominated_row = r1
                    print(f"Eliminated P1's {row_labels[r1]} (dominated by {row_labels[r2]})")
                    break
            if dominated_row is not None: break
        if dominated_row is not None:
            surviving_rows.remove(dominated_row)
            elimination_made = True
            continue

        dominated_col = None
        for c1 in surviving_cols:
            for c2 in surviving_cols:
                if c1 != c2 and all(matrix[r][c1][1] < matrix[r][c2][1] for r in surviving_rows):
                    dominated_col = c1
                    print(f"Eliminated P2's {col_labels[c1]} (dominated by {col_labels[c2]})")
                    break
            if dominated_col is not None: break
        if dominated_col is not None:
            surviving_cols.remove(dominated_col)
            elimination_made = True

    print(f"\nSurviving: P1 -> {[row_labels[r] for r in surviving_rows]}, P2 -> {[col_labels[c] for c in surviving_cols]}")
    if len(surviving_rows) == 1 and len(surviving_cols) == 1:
        print(f"Predicted Outcome: ({row_labels[surviving_rows[0]]}, {col_labels[surviving_cols[0]]})")

slide_matrix_3x4 = [
    [(1, 1), (-1, 2), (5, 0), (1, 1)],
    [(2, 3), (1, 2),  (3, 0), (5, 1)],
    [(1, 1), (0, 5),  (1, 7), (0, 1)]
]
iesds_solver(slide_matrix_3x4, ['A', 'B', 'C'], ['D', 'E', 'F', 'G'])
```

```python
# Q4 :

def find_pure_nash_equilibria(matrix, row_labels=None, col_labels=None):
    num_rows = len(matrix)
    num_cols = len(matrix[0]) if num_rows > 0 else 0
    if row_labels is None: row_labels = [f"Row {i}" for i in range(num_rows)]
    if col_labels is None: col_labels = [f"Col {j}" for j in range(num_cols)]

    p1_br, p2_br = set(), set()
    for c in range(num_cols):
        mx = max(matrix[r][c][0] for r in range(num_rows))
        for r in range(num_rows):
            if matrix[r][c][0] == mx: p1_br.add((r, c))
    for r in range(num_rows):
        mx = max(matrix[r][c][1] for c in range(num_cols))
        for c in range(num_cols):
            if matrix[r][c][1] == mx: p2_br.add((r, c))

    ne = p1_br.intersection(p2_br)
    if ne:
        for (r, c) in ne: print(f"Nash Equilibrium: ({row_labels[r]}, {col_labels[c]}) -> Payoff: {matrix[r][c]}")
    else:
        print("No pure-strategy Nash Equilibrium found.")

games = {
    "Prisoner's Dilemma": ([[(-1, -1), (-3, 0)], [(0, -3), (-2, -2)]], ['Cooperate', 'Defect'], ['Cooperate', 'Defect']),
    "Stag Hunt": ([[(4, 4), (0, 2)], [(2, 0), (2, 2)]], ['Stag', 'Hare'], ['Stag', 'Hare']),
    "Chicken": ([[(0, 0), (-1, 1)], [(1, -1), (-10, -10)]], ['Swerve', 'Straight'], ['Swerve', 'Straight']),
    "Battle of the Sexes": ([[(2, 1), (0, 0)], [(0, 0), (1, 2)]], ['Opera', 'Football'], ['Opera', 'Football']),
    "3x3 Slide Matrix": ([[(4, 4), (0, 0), (1, 3)], [(0, 0), (5, 5), (0, 0)], [(3, 1), (0, 0), (6, 6)]], ['A', 'B', 'C'], ['X', 'Y', 'Z'])
}
for name, (mat, r_l, c_l) in games.items():
    print(f"\n=== {name} ===")
    find_pure_nash_equilibria(mat, r_l, c_l)
```

```python
# Q5 :
import itertools

def find_n_player_nash(strategies, payoffs):
    all_profiles = list(itertools.product(*strategies))
    nash_equilibria = []
    for profile in all_profiles:
        is_nash = True
        for p_idx in range(len(strategies)):
            curr_payoff = payoffs[profile][p_idx]
            for alt_strat in strategies[p_idx]:
                if alt_strat == profile[p_idx]: continue
                mutated = list(profile)
                mutated[p_idx] = alt_strat
                if payoffs[tuple(mutated)][p_idx] > curr_payoff:
                    is_nash = False
                    break
            if not is_nash: break
        if is_nash: nash_equilibria.append(profile)
    for eq in nash_equilibria: print(f"Nash Equilibrium: {eq} -> Payoffs: {payoffs[eq]}")

three_firm_payoffs = {
    ('C', 'C', 'C'): [4, 4, 4], ('C', 'C', 'D'): [1, 1, 6], ('C', 'D', 'C'): [1, 6, 1], ('C', 'D', 'D'): [0, 3, 3],
    ('D', 'C', 'C'): [6, 1, 1], ('D', 'C', 'D'): [3, 0, 3], ('D', 'D', 'C'): [3, 3, 0], ('D', 'D', 'D'): [2, 2, 2]
}
find_n_player_nash([['C', 'D'], ['C', 'D'], ['C', 'D']], three_firm_payoffs)

# The unique Nash Equilibrium is ('D', 'D', 'D') because individual firms face an incentive structure to defect protectively.
# This represents a situatoin, where independent rational actions result in a suboptimal collective outcome
# compared to universal cooperation ('C', 'C', 'C').
```

```python
def interactive_game_solver():
    num_rows = int(input("Enter strategies for P1 (Rows): "))
    num_cols = int(input("Enter strategies for P2 (Columns): "))
    matrix = []
    for r in range(num_rows):
        row = []
        for c in range(num_cols):
            val = input(f"Cell ({r},{c}) [p1 p2]: ").split()
            row.append((int(val[0]), int(val[1])))
        matrix.append(row)

    print("\nMatrix Input:", matrix)

    print("\nDominant Strategies :")
    find_dominant_strategies(matrix)

    print("\nIESDS :")
    iesds_solver(matrix)

    print("\nPure Nash Equilibria :")
    find_pure_nash_equilibria(matrix)

interactive_game_solver()
```
