```python
# Question 2
def is_palindrome(s):
    processed_s = ''.join(char.lower() for char in s if char.isalnum())
    return processed_s == processed_s[::-1]
```

```python
# Question 3
def find_pair_with_diff(nums, x):
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if abs(nums[i] - nums[j]) == x:
                print(f"({nums[i]}, {nums[j]})")
                return
```

```python
# Question 4
import pandas as pd

data = {
    'Ticker': ['RELIANCE', 'TCS', 'INFY', 'HDFCBANK', 'WIPRO'],
    'Price': [2850.5, 3900.0, 1500.75, 1650.0, 420.3],
    'Volume': [1200000, 850000, 1100000, 950000, 700000],
    'Sector': ['Energy', 'IT', 'IT', 'Finance', 'IT']
}

df = pd.DataFrame(data)

print("Shape of the Dataframe:", df.shape)
print("\nFirst 3 rows:")
print(df.head(3))



print("Rows where Sector is 'IT':")
it_rows = df[df['Sector'] == 'IT']
print(it_rows)


df['Market_Cap_Proxy'] = df['Price'] * df['Volume']
print("Dataframe after adding 'Market_Cap_Proxy':")
print(df)


print("Dataframe sorted by Market_Cap_Proxy (Descending):")
sorted_df = df.sort_values(by='Market_Cap_Proxy', ascending=False)
print(sorted_df)
```

```python
# Question 6
import numpy as np

scores = np.array([45, 82, 76, 55, 90, 38, 67, 71])

mean_score = np.mean(scores)
median_score = np.median(scores)
std_dev = np.std(scores)

print(f"Mean: {mean_score}")
print(f"Median: {median_score}")
print(f"Standard Deviation: {std_dev:.2f}")


# Formula:(x - min)/(max - min)
normalized_scores = (scores - np.min(scores)) / (np.max(scores) - np.min(scores))

print("Normalized scores:")
print(normalized_scores)

above_mean_indices = np.where(scores > mean_score)[0]
print("Indices of students scoring above the mean:")
print(above_mean_indices)
```

```python
# Question 7
import matplotlib.pyplot as plt

mean = 0.0008
std_dev = 0.015
daily_returns = np.random.normal(loc=mean, scale=std_dev, size=252)


price_series = 100 * np.cumprod(1 + daily_returns)
price_series = np.insert(price_series, 0, 100.0)

plt.figure(figsize=(10, 5))
plt.plot(price_series, color='royalblue', linewidth=2, label='Stock Price')
plt.title('Simulated Stock Price Series Over 252 Days')
plt.xlabel('Trading Days')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()



final_price = price_series[-1]
peak_price = np.max(price_series)



running_peaks = np.maximum.accumulate(price_series)

drawdowns = (price_series - running_peaks) / running_peaks

max_drawdown = np.min(drawdowns)


print(f"Final Price:       {final_price:.2f}")
print(f"Peak Price:        {peak_price:.2f}")
print(f"Maximum Drawdown:  {max_drawdown * 100:.2f}%") # Displayed as a percentage
```

```python
# Question 8
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
year1 = [42, 45, 50, 48, 55, 60, 58, 63, 67, 70, 75, 80]
year2 = [52, 49, 58, 61, 66, 71, 68, 74, 79, 83, 88, 95]

y1_arr = np.array(year1)
y2_arr = np.array(year2)
difference = y2_arr - y1_arr

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(11, 10), sharex=False)

ax1.plot(months, year1, marker='o', color='royalblue', linewidth=2, label='Year 1')
ax1.plot(months, year2, marker='s', color='darkorange', linewidth=2, label='Year 2')

ax1.set_title('Monthly Revenue Comparison (Two Years)', fontsize=14, pad=15)
ax1.set_ylabel('Revenue (in lakhs)', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.legend(fontsize=11)


bar_colors = ['green' if diff >= 0 else 'red' for diff in difference]

ax2.bar(months, difference, color=bar_colors, edgecolor='black', alpha=0.8)

ax2.axhline(0, color='black', linewidth=1, linestyle='-')

ax2.set_title('Month-by-Month Revenue Difference (Year 2 - Year 1)', fontsize=14, pad=15)
ax2.set_xlabel('Months', fontsize=12)
ax2.set_ylabel('Difference (in lakhs)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Question 9
df = pd.DataFrame({
    'Day': range(1, 11),
    'Price': [100, 103, np.nan, 108, 107, np.nan, 112, 115, np.nan, 120],
    'Volume': [500, 520, 510, np.nan, 530, 545, np.nan, 560, 570, 580]
})

missing_counts = df.isnull().sum()
print(missing_counts)
print("\n" + "="*50 + "\n")


df['Price'] = df['Price'].ffill()

volume_mean = df['Volume'].mean()
df['Volume'] = df['Volume'].fillna(volume_mean)

print(df)
print("\n" + "="*50 + "\n")

df['Return'] = ((df['Price'] - df['Price'].shift(1)) / df['Price'].shift(1)) * 100

print(df)
```

```python
# Question 10
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
returns = np.random.multivariate_normal(
    mean=[0.0005, 0.0008, 0.0003],
    cov=[[0.0004,  0.0001,  0.00005],
         [0.0001,  0.0006,  0.00008],
         [0.00005, 0.00008, 0.0003]],
    size=200
)
prices = 100 * np.cumprod(1 + returns, axis=0)

df_prices = pd.DataFrame(prices, columns=['Asset_A', 'Asset_B', 'Asset_C'])

df_returns = pd.DataFrame(returns, columns=['Asset_A', 'Asset_B', 'Asset_C'])
correlation_matrix = df_returns.corr()


print(correlation_matrix)
print("\n" + "="*60 + "\n")

df_prices['Portfolio'] = df_prices[['Asset_A', 'Asset_B', 'Asset_C']].mean(axis=1)

plt.figure(figsize=(12, 6))

plt.plot(df_prices['Asset_A'], label='Asset A', linestyle='--', alpha=0.7)
plt.plot(df_prices['Asset_B'], label='Asset B', linestyle='--', alpha=0.7)
plt.plot(df_prices['Asset_C'], label='Asset C', linestyle='--', alpha=0.7)

plt.plot(df_prices['Portfolio'], label='Equally-Weighted Portfolio (1/3 Each)', color='black', linewidth=2.5)

plt.title('Asset Prices vs Equally-Weighted Portfolio Performance Over 200 Days', fontsize=14, pad=15)
plt.xlabel('Trading Days', fontsize=12)
plt.ylabel('Price Value', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11, loc='upper left')

plt.tight_layout()
plt.show()
```
